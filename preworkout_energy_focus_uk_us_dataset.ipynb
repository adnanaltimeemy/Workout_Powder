```python
import pandas as pd

# 1) Define the schema with example data
data = [
    {
        "country": "UK",
        "product_id": "PW-BR-001",
        "product_name": "BrandX Blue Raspberry Energy+Focus",
        "flavor": "Blue Raspberry",
        "variant": "300g tub",
        "year": 2025,
        "month": 1,
        "sales_channel": "Online",
        "region": "England",
        "units_sold": 500,
        "sell_price_per_unit": 35.00,
        "revenue": 17500.00,
        "cost_of_goods_sold": 7000.00,
        "marketing_cost": 1200.00,
        "fulfillment_cost": 800.00,
        # total_cost, gross_profit, net_profit will be computed
        "age_18_24_pct": 25,
        "age_25_34_pct": 40,
        "age_35_44_pct": 20,
        "age_45_54_pct": 10,
        "age_55_plus_pct": 5,
        "gender_male_pct": 70,
        "gender_female_pct": 28,
        "gender_other_pct": 2,
        "top_three_purchase_motivations": "Energy;Focus;Taste"
    },
    {
        "country": "US",
        "product_id": "PW-BR-001",
        "product_name": "BrandX Blue Raspberry Energy+Focus",
        "flavor": "Blue Raspberry",
        "variant": "300g tub",
        "year": 2025,
        "month": 1,
        "sales_channel": "Retail",
        "region": "Midwest",
        "units_sold": 800,
        "sell_price_per_unit": 39.00,
        "revenue": 31200.00,
        "cost_of_goods_sold": 12800.00,
        "marketing_cost": 1500.00,
        "fulfillment_cost": 1200.00,
        "age_18_24_pct": 18,
        "age_25_34_pct": 45,
        "age_35_44_pct": 22,
        "age_45_54_pct": 10,
        "age_55_plus_pct": 5,
        "gender_male_pct": 65,
        "gender_female_pct": 32,
        "gender_other_pct": 3,
        "top_three_purchase_motivations": "Energy;Focus;Brand"
    },
]

# 2) Create DataFrame
df = pd.DataFrame(data)

# 3) Derive additional financial fields
# total cost example: COGS + marketing + fulfillment
df["total_cost"] = df["cost_of_goods_sold"] + df["marketing_cost"] + df["fulfillment_cost"]

# gross profit: revenue - COGS
df["gross_profit"] = df["revenue"] - df["cost_of_goods_sold"]

# gross margin %
df["gross_margin_pct"] = df["gross_profit"] / df["revenue"] * 100

# net profit: revenue - total_cost
df["net_profit"] = df["revenue"] - df["total_cost"]

# net margin %
df["net_margin_pct"] = df["net_profit"] / df["revenue"] * 100

# 4) Reorder columns for readability (optional)
columns_order = [
    "country", "product_id", "product_name", "flavor", "variant",
    "year", "month", "sales_channel", "region",
    "units_sold", "sell_price_per_unit", "revenue",
    "cost_of_goods_sold", "marketing_cost", "fulfillment_cost",
    "total_cost", "gross_profit", "gross_margin_pct",
    "net_profit", "net_margin_pct",
    "age_18_24_pct", "age_25_34_pct", "age_35_44_pct",
    "age_45_54_pct", "age_55_plus_pct",
    "gender_male_pct", "gender_female_pct", "gender_other_pct",
    "top_three_purchase_motivations"
]
df = df[columns_order]

# 5) Export to CSV
output_file = "preworkout_dataset.csv"
df.to_csv(output_file, index=False)

print(f"CSV written to {output_file}")
print("\nPreview:")
print(df.head())
```
