```python
import pandas as pd

# 1) define schema columns
columns = [
    'region', 'date', 'product_name',
    'units_sold', 'unit_price', 'revenue',
    'cost_of_goods_sold', 'gross_profit', 'gross_margin_pct',
    'marketing_cost', 'net_profit', 'net_margin_pct',
    'age_18_24_pct', 'age_25_34_pct', 'age_35_44_pct', 'age_45_plus_pct',
    'use_sports_performance_pct', 'use_weight_loss_pct',
    'use_athletic_training_pct', 'use_recreational_fitness_pct',
    'notes'
]

# 2) example mock data row for UK
row_uk = {
    'region': 'UK',
    'date': pd.Timestamp(2026, 2, 28),
    'product_name': 'Kingslayer Pre Workout – Frozen Blue Ice',
    'units_sold': 120,
    'unit_price': 34.99,
    'revenue': 120 * 34.99,
    'cost_of_goods_sold': 120 * 15.00,
    'marketing_cost': 400.00,
    'age_18_24_pct': 20.0,
    'age_25_34_pct': 30.0,
    'age_35_44_pct': 25.0,
    'age_45_plus_pct': 25.0,
    'use_sports_performance_pct': 40.0,
    'use_weight_loss_pct': 20.0,
    'use_athletic_training_pct': 25.0,
    'use_recreational_fitness_pct': 15.0,
    'notes': 'Sample values; replace with real data.'
}

# Compute derived columns for UK
row_uk['gross_profit'] = row_uk['revenue'] - row_uk['cost_of_goods_sold']
row_uk['gross_margin_pct'] = (
    (row_uk['gross_profit'] / row_uk['revenue']) * 100
    if row_uk['revenue'] else 0
)
row_uk['net_profit'] = row_uk['gross_profit'] - row_uk['marketing_cost']
row_uk['net_margin_pct'] = (
    (row_uk['net_profit'] / row_uk['revenue']) * 100
    if row_uk['revenue'] else 0
)

# 3) example mock data row for US
row_us = {
    'region': 'US',
    'date': pd.Timestamp(2026, 2, 28),
    'product_name': 'Kingslayer Pre Workout – Frozen Blue Ice',
    'units_sold': 75,
    'unit_price': 59.99,
    'revenue': 75 * 59.99,
    'cost_of_goods_sold': 75 * 22.00,
    'marketing_cost': 350.00,
    'age_18_24_pct': 18.0,
    'age_25_34_pct': 32.0,
    'age_35_44_pct': 30.0,
    'age_45_plus_pct': 20.0,
    'use_sports_performance_pct': 45.0,
    'use_weight_loss_pct': 18.0,
    'use_athletic_training_pct': 22.0,
    'use_recreational_fitness_pct': 15.0,
    'notes': 'Sample values; replace with real data.'
}

# Compute derived columns for US
row_us['gross_profit'] = row_us['revenue'] - row_us['cost_of_goods_sold']
row_us['gross_margin_pct'] = (
    (row_us['gross_profit'] / row_us['revenue']) * 100
    if row_us['revenue'] else 0
)
row_us['net_profit'] = row_us['gross_profit'] - row_us['marketing_cost']
row_us['net_margin_pct'] = (
    (row_us['net_profit'] / row_us['revenue']) * 100
    if row_us['revenue'] else 0
)

# 4) build DataFrame directly (no concat warning)
df = pd.DataFrame([row_uk, row_us], columns=columns)

# 5) validation function
def validate_percentages(row):
    age_total = (
        row['age_18_24_pct'] +
        row['age_25_34_pct'] +
        row['age_35_44_pct'] +
        row['age_45_plus_pct']
    )

    use_total = (
        row['use_sports_performance_pct'] +
        row['use_weight_loss_pct'] +
        row['use_athletic_training_pct'] +
        row['use_recreational_fitness_pct']
    )

    return age_total, use_total

# Run validation
for idx, row in df.iterrows():
    age_total, use_total = validate_percentages(row)
    print(f"Row {idx} age pct total: {age_total}, use pct total: {use_total}")

# 6) export to CSV
output_path = 'kingslayer_sales_demographics_uk_us.csv'
df.to_csv(output_path, index=False)

print(f"Exported to {output_path}")
```
