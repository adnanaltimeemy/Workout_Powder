```python
import pandas as pd
from datetime import datetime

# -----------------------------------------------------------------------------
# Helper to compute revenue and margins for a row
# -----------------------------------------------------------------------------
def compute_financials(df, currency):
    # revenue
    df['revenue_local'] = df['units_sold'] * df['unit_price_local']
    # gross profit
    df['gross_profit_local'] = df['revenue_local'] - df['cost_of_goods_sold_local']
    # gross margin %
    df['gross_margin_pct'] = df.apply(
        lambda r: 100 * r['gross_profit_local'] / r['revenue_local']
        if r['revenue_local'] else 0,
        axis=1
    )
    # net profit
    df['net_profit_local'] = df['gross_profit_local'] - df['marketing_cost_local']
    # net margin %
    df['net_margin_pct'] = df.apply(
        lambda r: 100 * r['net_profit_local'] / r['revenue_local']
        if r['revenue_local'] else 0,
        axis=1
    )
    # tidy up: round financials to 2 decimals, margins to 1 decimal
    money_cols = ['unit_price_local', 'revenue_local', 'cost_of_goods_sold_local',
                  'gross_profit_local', 'marketing_cost_local', 'net_profit_local']
    df[money_cols] = df[money_cols].round(2)
    df['gross_margin_pct'] = df['gross_margin_pct'].round(1)
    df['net_margin_pct'] = df['net_margin_pct'].round(1)
    return df

# -----------------------------------------------------------------------------
# Example: UK data (GBP)
# -----------------------------------------------------------------------------
data_uk = [
    {
        'date': '2026-02-01',
        'channel': 'Retail',
        'units_sold': 120,
        'unit_price_local': 18.99,   # example UK price; adjust per source
        'cost_of_goods_sold_local': 8.50 * 120,  # placeholder COGS per unit
        'marketing_cost_local': 150.0,  # placeholder
        'demog_age_18_24_pct': 25.0,
        'demog_age_25_34_pct': 45.0,
        'demog_age_35_44_pct': 20.0,
        'demog_age_45_54_pct': 7.0,
        'demog_age_55_plus_pct': 3.0,
        'demog_male_pct': 70.0,
        'demog_female_pct': 28.0,
        'demog_other_pct': 2.0,
        'notes': 'Mixed Berry flavor; normal retail'
    },
    {
        'date': '2026-02-01',
        'channel': 'Ecommerce',
        'units_sold': 80,
        'unit_price_local': 18.99,
        'cost_of_goods_sold_local': 8.50 * 80,
        'marketing_cost_local': 120.0,
        'demog_age_18_24_pct': 30.0,
        'demog_age_25_34_pct': 40.0,
        'demog_age_35_44_pct': 20.0,
        'demog_age_45_54_pct': 5.0,
        'demog_age_55_plus_pct': 5.0,
        'demog_male_pct': 75.0,
        'demog_female_pct': 23.0,
        'demog_other_pct': 2.0,
        'notes': 'Online sale period'
    }
]

df_uk = pd.DataFrame(data_uk)
df_uk = compute_financials(df_uk, currency='GBP')

# -----------------------------------------------------------------------------
# Example: US data (USD) — placeholders because US listings may vary or be absent
# -----------------------------------------------------------------------------
data_us = [
    {
        'date': '2026-02-01',
        'channel': 'Retail',
        'units_sold': 50,
        'unit_price_local': 24.99,   # placeholder US price
        'cost_of_goods_sold_local': 11.00 * 50,  # placeholder COGS
        'marketing_cost_local': 200.0,
        'demog_age_18_24_pct': 20.0,
        'demog_age_25_34_pct': 50.0,
        'demog_age_35_44_pct': 20.0,
        'demog_age_45_54_pct': 7.0,
        'demog_age_55_plus_pct': 3.0,
        'demog_male_pct': 68.0,
        'demog_female_pct': 30.0,
        'demog_other_pct': 2.0,
        'notes': 'Hypothetical US retail'
    },
    {
        'date': '2026-02-01',
        'channel': 'Ecommerce',
        'units_sold': 70,
        'unit_price_local': 24.99,
        'cost_of_goods_sold_local': 11.00 * 70,
        'marketing_cost_local': 180.0,
        'demog_age_18_24_pct': 25.0,
        'demog_age_25_34_pct': 45.0,
        'demog_age_35_44_pct': 20.0,
        'demog_age_45_54_pct': 7.0,
        'demog_age_55_plus_pct': 3.0,
        'demog_male_pct': 72.0,
        'demog_female_pct': 26.0,
        'demog_other_pct': 2.0,
        'notes': 'Hypothetical US online'
    }
]

df_us = pd.DataFrame(data_us)
df_us = compute_financials(df_us, currency='USD')

# -----------------------------------------------------------------------------
# Output to CSV
# -----------------------------------------------------------------------------
df_uk.to_csv('warrior_creatine_mixberry_uk.csv', index=False)
df_us.to_csv('warrior_creatine_mixberry_us.csv', index=False)

print("UK and US CSVs created with sample data.")
```
